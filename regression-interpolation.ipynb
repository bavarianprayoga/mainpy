```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
```

Least Square Regression

alpha[0] = m
alpha[1] = b

Beta = (A^T * A)^-1 * A^T * Y

```python
x = [1, 2, 3, 4, 5, 6, 7, 8]
y =[7, 5, 4, 3, 7, 8, 11, 6]

x = np.array(x) # convert list to numpy array
y = np.array(y) # convert list to numpy array

A = np.vstack([x, np.ones(len(x))]).T # add a column of 1's to A, T = transpose
y = y[:, np.newaxis] # slice Y into a column vector, newaxis = increase the dimension of the existing array by one more dimension

beta = np.dot(np.dot(np.linalg.inv(np.dot(A.T, A)), A.T), y) # ini rumus yang Beta = (A^T * A)^-1 * A^T * Y
m = beta[0]
b = beta[1]

# Cara Hafal
#
# ATAATY
# Bagi 3, 2, 1 -> ATA AT Y
# np.dot -> np.dot -> np.linalg.inv -> np.dot
# 
# Masukkan satu per satu
# 1. np.dot(, Y)
# 2. np.dot(np.dot(, A.T), Y)
# 3. np.dot(np.dot(np.linalg.inv(), A.T), Y)
# 4. np.dot(np.dot(np.lingalg.inv(np.dot(A.T, A)), A.T), Y)

print(A)
print(y)
print(beta)

# 5. Plotting Garis
# --------------------------------------------
#    Masukin ke rumus y = mx + b
#    Parameter dari plotting (sumbu x, sumbu y, styling)

plt.plot(x, y, 'ro') # plot the data
plt.plot(x, m*x + b, 'red', label='Gatau') # plot the line
plt.legend()
plt.show()  
```

Polynomial Regression

```python
x = [1, 2, 3, 4, 5, 6, 7, 8]
y =[7, 5, 4, 3, 7, 8, 11, 6]

x = np.array(x)
y = np.array(y)

for i in range(1, 7):
    y_reg = np.polyfit(x, y, i) # ini rumus yang Beta = (A^T * A)^-1 * A^T * Y
    # print(y_reg)

    plt.subplot(2, 3, i)
    plt.plot(x, y, 'blue') # plot the data
    plt.plot(x, np.polyval(y_reg, x), 'red', label='Gatau') # plot the line
plt.legend()
plt.show()
```

Cubic Spline Interpolation

```python
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [7, 5, 4, 3, 7, 8, 11, 6]

x = np.array(x)
y = np.array(y)

f = CubicSpline(x, y, bc_type='natural') # interpolate the data with a cubic spline, bc_type = boundary condition
x_new = np.linspace(1, 8, 1000) # generate 100 points between 1 and 8
y_new = f(x_new) # evaluate the spline at the new points

plt.plot(x, y, 'ro') # plot the data
plt.plot(x_new, y_new, label='Gatau') # plot the line
plt.legend()
plt.show()
```
